```python
employees = pd.read_csv('employee_attrition_data.csv')
```

```python
# Limit the returns to numeric values to avoid errors in the map:

df_numeric = employees.select_dtypes(include='number')
corr_matrix = df_numeric.corr()

plt.figure(figsize=(10, 8))  # Optional: adjust the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap from Employees Attrition Data')
plt.show()
```
